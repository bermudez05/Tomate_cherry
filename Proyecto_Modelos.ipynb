Este tomate ha adquirido relevancia en la producción agrícola debido a su sabor dulce y alto valor nutricional, su demanda en el mercado y su adaptabilidad a diversas condiciones climáticas. Su producción es significativa en la agricultura intensiva y protegida, como los invernaderos, donde se pueden controlar mejor las variables ambientales.

#### 3.1.2 Condiciones Edafoclimáticas
Las condiciones edafoclimáticas se refieren a los factores ambientales que interactúan entre sí, que influyen en el crecimiento y desarrollo de las plantas .
En el crecimiento y desarrollo del tomate cherry las condiciones edafoclimáticas más importantes se encuentran:  

* **Temperatura:** La temperatura influye directamente en la fotosíntesis, la respiración y la transpiración de la planta. La temperatura óptima para el crecimiento del tomate cherry oscila entre 18°C y 25°C, mientras que temperaturas por encima de 30°C pueden afectar la floración y el cuajado de frutos.  

* **Humedad relativa:** Un nivel adecuado de humedad relativa (entre 55% y 75%) es crucial para evitar problemas como el estrés hídrico y la proliferación de enfermedades fúngicas. Una humedad excesiva puede favorecer el desarrollo de patógenos, mientras que una humedad baja puede aumentar la transpiración y afectar la absorción de agua y nutrientes.  

* **pH del suelo:**  El pH del suelo afecta la disponibilidad de nutrientes para las plantas. El tomate cherry se desarrolla mejor en suelos con un pH entre 5.75 y 6.8, ya que fuera de este rango pueden presentarse deficiencias nutricionales o toxicidad por ciertos elementos.  

* **Exposición solar y fotosíntesis:**  El tomate cherry es una planta de alta demanda lumínica. La radiación solar es el principal factor que impulsa la fotosíntesis, proceso mediante el cual la planta sintetiza azúcares esenciales para su crecimiento. Se recomienda un mínimo de 8 horas de luz diaria para un desarrollo óptimo.  

* **Macronutrientes (N, P, K):**   Los macronutrientes esenciales para el tomate cherry incluyen:  
  - **Nitrógeno (N):** Es fundamental para el desarrollo vegetativo, promoviendo el crecimiento de hojas y tallos.  
  - **Fósforo (P):** Contribuye a la formación de raíces y flores, favoreciendo la producción de frutos.  
  - **Potasio (K):** Regula la absorción de agua, mejora la calidad de los frutos  y la resistencia a enfermedades.  

El equilibrio en la disponibilidad de estos nutrientes es esencial para obtener una producción óptima.

### 4.2 Modelos Matemáticos de Crecimiento de Cultivos  

Existen múltiples modelos matemáticos utilizados en agronomía para representar el crecimiento de los cultivos. Entre ellos, los modelos más comúnmente empleados incluyen:  

- **Modelos Sigmoideos** (Logístico y Gompertz), que representan el crecimiento con una curva en forma de "S", útil cuando hay factores limitantes en el desarrollo del cultivo.  
- **Modelos Expo-Lineales**, donde el crecimiento inicialmente es exponencial y luego se vuelve lineal, útil en entornos controlados.  
- **Modelos Basados en Biomasa**, que consideran la acumulación de materia seca en función de los recursos disponibles, como luz y agua.  

Para este estudio, hemos decidido **implementar y comparar los modelos de acumulación de biomasa y expo-lineal**, ya que nos permiten evaluar cómo las variables ambientales influyen directamente en la cantidad de biomasa producida por la planta.  

#### 4.2.1 Modelo de Acumulación de Biomasa

El **modelo de acumulación de biomasa** describe el crecimiento de las plantas en función de la **radiación fotosintéticamente activa (PAR)** y la **eficiencia en el uso de los recursos**. Es útil para analizar cómo la luz, la humedad del suelo y la temperatura afectan la producción de biomasa.  

En este proyecto, implementaremos la siguiente ecuación diferencial ordinaria (EDO) para modelar la acumulación de biomasa del tomate cherry:  

$$
\frac{dB}{dt} = \alpha R B - \beta B
$$  

Donde:  
- $B(t)$ es la biomasa acumulada en el tiempo.  
- $\alpha$ es la tasa de conversión de recursos en biomasa.  
- $\beta$ es la tasa de pérdida de biomasa.  
- $R(T, H, C, L)$ es la función que representa la disponibilidad de recursos en función de la temperatura ($T$), humedad del suelo ($H$), conductividad eléctrica ($C$) e intensidad lumínica ($L$).  

Para definir $R$, usamos la siguiente función lineal:  

$$
R(T, H, C, L) = k_1 T + k_2 H + k_3 C + k_4 L
$$  

Donde $k_1, k_2, k_3, k_4$ son coeficientes que ponderan la influencia de cada factor en la acumulación de biomasa.  

##### **Explicación de la EDO en el Contexto del Proyecto**:

Esta ecuación diferencial describe el crecimiento de la biomasa en términos de **una tasa de acumulación neta**. La tasa de cambio de la biomasa $\frac{dB}{dt}$ depende de dos términos:  

1. **Producción de biomasa:** $\alpha R B$ → Representa el crecimiento proporcional a la disponibilidad de recursos ($R$) y la biomasa existente ($B$).  
2. **Pérdida de biomasa:** $\beta B$ → Representa la biomasa perdida debido a procesos naturales como respiración celular y estrés ambiental.  

La ventaja de este modelo es que permite evaluar cómo cada variable ambiental ($T, H, C, L$) contribuye a la producción total de biomasa, lo que lo hace altamente interpretable para la optimización del cultivo.  

#### 4.2.2 Modelo Expo-Lineal

El **modelo expo-lineal** es útil en cultivos donde el crecimiento sigue dos fases distintas:  

1. **Fase exponencial:** En las primeras etapas de crecimiento, cuando la planta tiene acceso a recursos abundantes, el crecimiento es rápido y se modela mediante una función exponencial.  
2. **Fase lineal:** A medida que el crecimiento avanza, factores limitantes como la disponibilidad de nutrientes y espacio reducen la tasa de crecimiento, que se estabiliza en un patrón lineal.  

Este modelo se describe mediante la siguiente ecuación diferencial ordinaria (EDO):  

$$
\frac{dB}{dt} =
\begin{cases}
\alpha R B, & t < t_s \quad (\text{fase exponencial})\\
\gamma R, & t \geq t_s \quad (\text{fase lineal})
\end{cases}
$$  

Donde:  
- $B(t)$ es la biomasa acumulada en el tiempo.  
- $\alpha$ es la tasa de crecimiento en la fase exponencial.  
- $\gamma$ es la tasa de crecimiento en la fase lineal.  
- $t_s$ es el tiempo de transición entre ambas fases.  
- $R(T, H, C, L)$ es la función que representa la disponibilidad de recursos en función de la temperatura ($T$), humedad del suelo ($H$), conductividad eléctrica ($C$) e intensidad lumínica ($L$).  

La función de recursos se define como:  

$$
R(T, H, C, L) = k_1 T + k_2 H + k_3 C + k_4 L
$$  

Donde $k_1, k_2, k_3, k_4$ son coeficientes que ponderan la influencia de cada factor en la acumulación de biomasa.  

##### **Explicación de la EDO en el Contexto del Proyecto:**

La ecuación diferencial describe el **crecimiento en dos fases**:  

1. **Fase exponencial ($t < t_s$):**  
   - El crecimiento depende de la biomasa acumulada y la disponibilidad de recursos.  
   - Representada por la ecuación:  
     $$
     \frac{dB}{dt} = \alpha R B
     $$  

2. **Fase lineal ($t \geq t_s$):**  
   - La tasa de crecimiento ya no depende de la biomasa acumulada, sino solo de la disponibilidad de recursos.  
   - Se estabiliza a un ritmo constante.  
   - Representada por la ecuación:  
     $$
     \frac{dB}{dt} = \gamma R
     $$  

Esta formulación permite evaluar **cómo la transición entre ambas fases es influenciada por las condiciones ambientales**, lo que es clave en la optimización del cultivo de tomate cherry.

### 5.2 Implementación modelo de acumulación de Biomasa


```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

Ambas implementaciones de los modelos comparten estos parametros

```python
# Datos de entrada
T = cherry_tomato['temperature']  # Temperatura en °C
M = cherry_tomato['moisture']          # Humedad del suelo en %
C = cherry_tomato['conductivity']  # Conductividad eléctrica en µS/cm
L = cherry_tomato['light']  # Intensidad de luz en lx

# Definimos la función de recursos
# Calcula el factor de recursos R en función de las variables ambientales.
def recursos(T, M, C, L, k1, k2, k3, k4):
    R = k1 * T + k2 * M + k3 * C + k4 * L
    return R
```

```python
# Definimos el modelo de acumulación de biomasa como se planteo inicialmente
def biomass_growth(t, B, alpha, beta, k1, k2, k3, k4):
    # Calculamos R en función del tiempo
    idx = int(t)  # Suponemos que t es un índice de tiempo
    R = recursos(T[idx], M[idx], C[idx], L[idx], k1, k2, k3, k4)
    return alpha * R * B - beta * B


# Parámetros iniciales los cuales requieren ser ajustados
alpha = 0.1  # Tasa de conversión de recursos en biomasa
beta = 0.02  # Tasa de pérdida de biomasa
k1, k2, k3, k4 = 0.1, 0.2, 0.05, 0.15  # Coeficientes de la función de recursos


# Condición inicial
B0 = [1.0]  # Dado que nuestro marco de datos no tiene un dataset asignamos una biomasa inicial

# Intervalo de tiempo (usamos el índice de los datos)
t_span = (0, len(T) - 1)  # Desde el primer hasta el último registro

# Resolver la EDO
sol = solve_ivp(biomass_growth, t_span, B0, args=(alpha, beta, k1, k2, k3, k4), t_eval=np.arange(len(T)))

# Graficar la solución
plt.plot(sol.t, sol.y[0], label='Biomasa simulada')
plt.xlabel('Tiempo ')
plt.ylabel('Biomasa')
plt.title('Simulación del crecimiento de biomasa')
plt.legend()
plt.show()
```

### 5.3 Implementación modelo Expo-Lineal

```python
# Función de recursos
def recursos(T, M, C, L, k1, k2, k3, k4):
    R = k1 * T + k2 * M + k3 * C + k4 * L
    return R

def expo_linear_growth(t, B, alpha, gamma, t_switch, k1, k2, k3, k4):
    idx = min(int(t), len(T) - 1)  # Dentro de los limites del índice
    R = recursos(T[idx], M[idx], C[idx], L[idx], k1, k2, k3, k4)

    if t < t_switch:
        return alpha * R * B  # Fase exponencial (dB/dt = alpha * R * B)
    else:
        return gamma * R      # Fase lineal (dB/dt = gamma * R)
```

Para este modelo expolineal definiremos los siguientes parametros, los cuales impactan severamente en las predicciones. Los cuales son suceptibles a mejora.

```python
# Parámetros iniciales
alpha = 0.2  # Tasa de crecimiento exponencial
gamma = 0.05  # Tasa de crecimiento lineal
t_switch = 3  # Tiempo de transición de un comportamiento exponencial a lineal (índice)
k1, k2, k3, k4 = 0.1, 0.2, 0.05, 0.15  # Coeficientes de la función de recursos

# Los coeficientes de la función de recursos se ajustan de acuerdo a las descripciones
# de la incidencia de las variables en el crecimiento del tomate cherry

# Condición de biomasa inicial
B0 = [1.0]

# Intervalo de tiempo
t_span = (0, len(T) - 1)
```

Finalmente hacemos la resolución del modelo de EDO planteado

```python
# Resolución de la la EDO planteada
sol = solve_ivp(expo_linear_growth, t_span, B0, args=(alpha, gamma, t_switch, k1, k2, k3, k4), t_eval=np.arange(len(T)))

# Solución grafica
plt.plot(sol.t, sol.y[0], label='Biomasa simulada')
plt.xlabel('Tiempo')
plt.ylabel('Biomasa')
plt.title('Simulación del crecimiento expo-lineal')
plt.legend()
plt.show()
```

#### 5.3.1 Simulación de datos

Dado que no contamos con medidas reales de la biomasa en tomates cherrys, generaremos un conjunto de datos sinteticos basados en el modelo expo-lineal que se ajusta más a las descripciones descritas <a href="https://bio.libretexts.org/Bookshelves/Botany/Inanimate_Life_(Briggs)/01%3A_Chapters/1.25%3A_Plant_growthpatterns_limitations_and_models#title">Plant Growth Patterns , Limitations, and Models</a>. Este enfoque (expo-lineal) considera que las plantas atraviesan una fase de crecimiento exponencial seguida de una fase de crecimiento lineal, lo que lo hace adecuado para representar el desarrollo de la biomasa del tomate cherry.  

```python
# Generaramos datos sintéticos basados en el modelo propuesto expo-lineal (se asemeja al comportamiento descrito del crecimiento del Tomate Cherry)
def generar_datos_sinteticos(T, M, C, L, alpha, gamma, t_switch, k1, k2, k3, k4):
    t_span = (0, len(T) - 1)
    t_eval = np.arange(len(T))
    sol = solve_ivp(expo_linear_growth, t_span, B0, args=(alpha, gamma, t_switch, k1, k2, k3, k4), t_eval=np.arange(len(T)))
    return sol.t, sol.y[0]


def agr_ruido(datos, sigma=0.05):
    ruido = np.random.normal(0, sigma, size=len(datos))
    return datos + ruido
```

```python
data_sim = generar_datos_sinteticos(T,M,C,L,alpha,gamma,t_switch,k1,k2,k3,k4) # Retornamos tiempo y datos
data_ruido = agr_ruido(data_sim[1])
```

```python
def perturbar_datos(datos, intervalo=20, intensidad=0.3):
    datos_perturbados = datos.copy()
    for i in range(0, len(datos), intervalo):
        datos_perturbados[i] += np.random.uniform(-intensidad, intensidad)
    return datos_perturbados

data_pert = perturbar_datos(data_ruido,20,2.2)

plt.plot(data_sim[0], data_pert, label='Biomasa simulada')
plt.xlabel('Tiempo ')
plt.ylabel('Biomasa')
plt.title('Simulación del crecimiento expo-lineal')
plt.legend()
plt.show()
```

#### 5.3.2 Ajuste de parametros

Optimizamos para encontrar los parámetros que mejor ajusten la simulación a los datos perturbados. Minimizamos la diferencia cuadrática entre los valores simulados y los valores perturbados y comparamos los datos ajustados con los datos perturbados para evaluar el rendimiento del modelo.

```python
# Definir la función de residuo (error cuadrático entre la simulación y los datos perturbados)
def residuo(parametros, T, M, C, L, datos_perturbados):
    # Extraer parámetros
    alpha, gamma, t_switch, k1, k2, k3, k4 = parametros

    # Simular los datos con los parámetros actuales
    t_span = (0, len(T) - 1)
    sol = solve_ivp(expo_linear_growth, t_span, [1.0], args=(alpha, gamma, t_switch, k1, k2, k3, k4), t_eval=np.arange(len(T)))

    # Comparar la simulación con los datos perturbados
    simulacion = sol.y[0]
    error = np.sum((simulacion - datos_perturbados)**2)  # Error cuadrático

    return error
```

```python
# Parámetros iniciales para la optimización
parametros_iniciales = [0.1, 0.05, 3, 0.1, 0.2, 0.05, 0.15]

# Optimización: minimizar el error cuadrático
resultado = minimize(residuo, parametros_iniciales, args=(T, M, C, L, data_pert), bounds=[(0, 3), (0, 3), (0, 20), (0, 3), (0, 3), (0, 3), (0, 3)],method='Powell')

# Parámetros ajustados
print("Parámetros ajustados:", resultado.x)

# Recuperar los parámetros ajustados
alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt = resultado.x

# Generar la simulación con los parámetros ajustados
sol = solve_ivp(expo_linear_growth, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt), t_eval=np.arange(len(T)))

# Graficar comparación entre los datos simulados perturbados y la simulación ajustada
plt.plot(data_sim[0],data_sim[1], label='Biomasa simulada')
plt.plot(sol.t, sol.y[0], label='Simulación de la biomasa ajustada')
plt.xlabel('Tiempo ')
plt.ylabel('Biomasa ')
plt.legend()
plt.title('Ajuste de parámetros al modelo expo-lineal con perturbaciones')
plt.show()
```

De este modo obtenemos el siguiente modelo con los parametros adecuados a los datos simulados.


\begin{align}
B(t) = 0.7086⋅RB, t < t_{switch}
\end{align}

\begin{align}

\end{align}

\begin{align}
R=0.0969T+0.2157M+0.5147C+0.8209L
\end{align}

Ahora lo que haremos sera un análisis sobre nuestro modelo para responder a ¿Qué tanto varia el crecimiento es diferentes etapas?

```python

def expo_linear_growth_mod(t, B, alpha, gamma, t_switch, k1, k2, k3, k4, T, M, C, L):
    idx = min(int(t), len(T) - 1)  # Dentro de los limites del índice
    R = recursos(T[idx], M[idx], C[idx], L[idx], k1, k2, k3, k4)

    if t < t_switch:
        return alpha * R * B  # Fase exponencial (dB/dt = alpha * R * B)
    else:
        return gamma * R

def aplicar_variacion(vector, porcentaje):
    """
    Aplica una variación porcentual a un vector.
    :param vector: Vector original (1 x N).
    :param porcentaje: Porcentaje de variación (positivo o negativo).
    :return: Vector modificado.
    """
    factor = 1 + porcentaje / 100  # Convertir porcentaje a factor multiplicativo
    return vector * factor

# Definir el porcentaje de variación deseado
porcentaje_variacion = 30

# Aplicar variaciones a cada variable
T_mas = aplicar_variacion(T, porcentaje_variacion)  # +30%
T_menos = aplicar_variacion(T, -porcentaje_variacion)  # -30%

M_mas = aplicar_variacion(M, porcentaje_variacion)  # +30%
M_menos = aplicar_variacion(M, -porcentaje_variacion)  # -30%

C_mas = aplicar_variacion(C, porcentaje_variacion)  # +30%
C_menos = aplicar_variacion(C, -porcentaje_variacion)  # -30%

L_mas = aplicar_variacion(L, porcentaje_variacion)  # +30%
L_menos = aplicar_variacion(L, -porcentaje_variacion)  # -30%

resultados = []
etiquetas = []

B0 = [1.0]
t_span = (0, len(T) - 1)

# Simulaciones para temperatura

sol_temp_mas = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T_mas, M, C, L), t_eval=np.arange(len(T)))
resultados.append(sol_temp_mas.y[0])
etiquetas.append(f'Temperatura +{porcentaje_variacion}%')

sol_temp_menos = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T_menos, M, C, L), t_eval=np.arange(len(T)))
resultados.append(sol_temp_menos.y[0])
etiquetas.append(f'Temperatura -{porcentaje_variacion}%')

# Simulaciones para humedad
sol_hum_mas = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T, M_mas, C, L), t_eval=np.arange(len(T)))
resultados.append(sol_hum_mas.y[0])
etiquetas.append(f'Humedad +{porcentaje_variacion}%')

sol_hum_menos = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T, M_menos, C, L), t_eval=np.arange(len(T)))
resultados.append(sol_hum_menos.y[0])
etiquetas.append(f'Humedad -{porcentaje_variacion}%')

# Simulaciones para conductividad eléctrica
sol_con_mas = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T, M, C_mas, L), t_eval=np.arange(len(T)))
resultados.append(sol_con_mas.y[0])
etiquetas.append(f'Conductividad +{porcentaje_variacion}%')

sol_con_menos = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt,  T, M, C_menos, L), t_eval=np.arange(len(T)))
resultados.append(sol_con_menos.y[0])
etiquetas.append(f'Conductividad -{porcentaje_variacion}%')

# Simulaciones para luminosidad solar
sol_lu_mas = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T, M, C, L_mas), t_eval=np.arange(len(T)))
resultados.append(sol_lu_mas.y[0])
etiquetas.append(f'Luminosidad +{porcentaje_variacion}%')

sol_lu_menos = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt,  T, M, C, L_menos), t_eval=np.arange(len(T)))
resultados.append(sol_lu_menos.y[0])
etiquetas.append(f'Luminosidad -{porcentaje_variacion}%')

# Graficar los resultados
plt.figure(figsize=(12, 8))
for i, resultado in enumerate(resultados):
    plt.plot(resultado, label=etiquetas[i])

plt.title('Impacto de las Variaciones Ambientales en el Crecimiento de la Biomasa')
plt.xlabel('Tiempo')
plt.ylabel('Biomasa')
plt.legend()
plt.grid(True)
plt.show()
```

Podemos concluir que segun nuestro modelo lo que tiene un impacto más positivo para el crecimiento es la humedad y la conductividad positiva asi como la buena luminosidad, ademas lo más perjudicial para el crecimiento son las bajas temperaturas y humedades, como la poca luminosidad
