### 4.2 Modelos Matem√°ticos de Crecimiento de Cultivos  

Existen m√∫ltiples modelos matem√°ticos utilizados en agronom√≠a para representar el crecimiento de los cultivos. Entre ellos, los modelos m√°s com√∫nmente empleados incluyen:  

- **Modelos Sigmoideos** (Log√≠stico y Gompertz), que representan el crecimiento con una curva en forma de "S", √∫til cuando hay factores limitantes en el desarrollo del cultivo.  
- **Modelos Expo-Lineales**, donde el crecimiento inicialmente es exponencial y luego se vuelve lineal, √∫til en entornos controlados.  
- **Modelos Basados en Biomasa**, que consideran la acumulaci√≥n de materia seca en funci√≥n de los recursos disponibles, como luz y agua.  

Para este estudio, hemos decidido **implementar y comparar los modelos de acumulaci√≥n de biomasa y expo-lineal**, ya que nos permiten evaluar c√≥mo las variables ambientales influyen directamente en la cantidad de biomasa producida por la planta.  

#### 4.2.1 Modelo de Acumulaci√≥n de Biomasa

El **modelo de acumulaci√≥n de biomasa** describe el crecimiento de las plantas en funci√≥n de la **radiaci√≥n fotosint√©ticamente activa (PAR)** y la **eficiencia en el uso de los recursos**. Es √∫til para analizar c√≥mo la luz, la humedad del suelo y la temperatura afectan la producci√≥n de biomasa.  

En este proyecto, implementaremos la siguiente ecuaci√≥n diferencial ordinaria (EDO) para modelar la acumulaci√≥n de biomasa del tomate cherry:  

$$
\frac{dB}{dt} = \alpha R B - \beta B
$$  

Donde:  
- $B(t)$ es la biomasa acumulada en el tiempo.  
- $\alpha$ es la tasa de conversi√≥n de recursos en biomasa.  
- $\beta$ es la tasa de p√©rdida de biomasa.  
- $R(T, H, C, L)$ es la funci√≥n que representa la disponibilidad de recursos en funci√≥n de la temperatura ($T$), humedad del suelo ($H$), conductividad el√©ctrica ($C$) e intensidad lum√≠nica ($L$).  

Para definir $R$, usamos la siguiente funci√≥n lineal:  

$$
R(T, H, C, L) = k_1 T + k_2 H + k_3 C + k_4 L
$$  

Donde $k_1, k_2, k_3, k_4$ son coeficientes que ponderan la influencia de cada factor en la acumulaci√≥n de biomasa.  

##### **Explicaci√≥n de la EDO en el Contexto del Proyecto**:

Esta ecuaci√≥n diferencial describe el crecimiento de la biomasa en t√©rminos de **una tasa de acumulaci√≥n neta**. La tasa de cambio de la biomasa $\frac{dB}{dt}$ depende de dos t√©rminos:  

1. **Producci√≥n de biomasa:** $\alpha R B$ ‚Üí Representa el crecimiento proporcional a la disponibilidad de recursos ($R$) y la biomasa existente ($B$).  
2. **P√©rdida de biomasa:** $\beta B$ ‚Üí Representa la biomasa perdida debido a procesos naturales como respiraci√≥n celular y estr√©s ambiental.  

La ventaja de este modelo es que permite evaluar c√≥mo cada variable ambiental ($T, H, C, L$) contribuye a la producci√≥n total de biomasa, lo que lo hace altamente interpretable para la optimizaci√≥n del cultivo.  

#### 4.2.2 Modelo Expo-Lineal

El **modelo expo-lineal** es √∫til en cultivos donde el crecimiento sigue dos fases distintas:  

1. **Fase exponencial:** En las primeras etapas de crecimiento, cuando la planta tiene acceso a recursos abundantes, el crecimiento es r√°pido y se modela mediante una funci√≥n exponencial.  
2. **Fase lineal:** A medida que el crecimiento avanza, factores limitantes como la disponibilidad de nutrientes y espacio reducen la tasa de crecimiento, que se estabiliza en un patr√≥n lineal.  

Este modelo se describe mediante la siguiente ecuaci√≥n diferencial ordinaria (EDO):  

$$
\frac{dB}{dt} =
\begin{cases}
\alpha R B, & t < t_s \quad (\text{fase exponencial})\\
\gamma R, & t \geq t_s \quad (\text{fase lineal})
\end{cases}
$$  

Donde:  
- $B(t)$ es la biomasa acumulada en el tiempo.  
- $\alpha$ es la tasa de crecimiento en la fase exponencial.  
- $\gamma$ es la tasa de crecimiento en la fase lineal.  
- $t_s$ es el tiempo de transici√≥n entre ambas fases.  
- $R(T, H, C, L)$ es la funci√≥n que representa la disponibilidad de recursos en funci√≥n de la temperatura ($T$), humedad del suelo ($H$), conductividad el√©ctrica ($C$) e intensidad lum√≠nica ($L$).  

La funci√≥n de recursos se define como:  

$$
R(T, H, C, L) = k_1 T + k_2 H + k_3 C + k_4 L
$$  

Donde $k_1, k_2, k_3, k_4$ son coeficientes que ponderan la influencia de cada factor en la acumulaci√≥n de biomasa.  

##### **Explicaci√≥n de la EDO en el Contexto del Proyecto:**

La ecuaci√≥n diferencial describe el **crecimiento en dos fases**:  

1. **Fase exponencial ($t < t_s$):**  
   - El crecimiento depende de la biomasa acumulada y la disponibilidad de recursos.  
   - Representada por la ecuaci√≥n:  
     $$
     \frac{dB}{dt} = \alpha R B
     $$  

2. **Fase lineal ($t \geq t_s$):**  
   - La tasa de crecimiento ya no depende de la biomasa acumulada, sino solo de la disponibilidad de recursos.  
   - Se estabiliza a un ritmo constante.  
   - Representada por la ecuaci√≥n:  
     $$
     \frac{dB}{dt} = \gamma R
     $$  

Esta formulaci√≥n permite evaluar **c√≥mo la transici√≥n entre ambas fases es influenciada por las condiciones ambientales**, lo que es clave en la optimizaci√≥n del cultivo de tomate cherry.

### 5.2 Implementaci√≥n modelo de acumulaci√≥n de Biomasa


```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

Ambas implementaciones de los modelos comparten estos parametros

```python
# Datos de entrada
T = cherry_tomato['temperature']  # Temperatura en ¬∞C
M = cherry_tomato['moisture']          # Humedad del suelo en %
C = cherry_tomato['conductivity']  # Conductividad el√©ctrica en ¬µS/cm
L = cherry_tomato['light']  # Intensidad de luz en lx

# Definimos la funci√≥n de recursos
# Calcula el factor de recursos R en funci√≥n de las variables ambientales.
def recursos(T, M, C, L, k1, k2, k3, k4):
    R = k1 * T + k2 * M + k3 * C + k4 * L
    return R
```

```python
# Definimos el modelo de acumulaci√≥n de biomasa como se planteo inicialmente
def biomass_growth(t, B, alpha, beta, k1, k2, k3, k4):
    # Calculamos R en funci√≥n del tiempo
    idx = int(t)  # Suponemos que t es un √≠ndice de tiempo
    R = recursos(T[idx], M[idx], C[idx], L[idx], k1, k2, k3, k4)
    return alpha * R * B - beta * B


# Par√°metros iniciales los cuales requieren ser ajustados
alpha = 0.1  # Tasa de conversi√≥n de recursos en biomasa
beta = 0.02  # Tasa de p√©rdida de biomasa
k1, k2, k3, k4 = 0.1, 0.2, 0.05, 0.15  # Coeficientes de la funci√≥n de recursos


# Condici√≥n inicial
B0 = [1.0]  # Dado que nuestro marco de datos no tiene un dataset asignamos una biomasa inicial

# Intervalo de tiempo (usamos el √≠ndice de los datos)
t_span = (0, len(T) - 1)  # Desde el primer hasta el √∫ltimo registro

# Resolver la EDO
sol = solve_ivp(biomass_growth, t_span, B0, args=(alpha, beta, k1, k2, k3, k4), t_eval=np.arange(len(T)))

# Graficar la soluci√≥n
plt.plot(sol.t, sol.y[0], label='Biomasa simulada')
plt.xlabel('Tiempo ')
plt.ylabel('Biomasa')
plt.title('Simulaci√≥n del crecimiento de biomasa')
plt.legend()
plt.show()
```

### 5.3 Implementaci√≥n modelo Expo-Lineal

```python
# Funci√≥n de recursos
def recursos(T, M, C, L, k1, k2, k3, k4):
    R = k1 * T + k2 * M + k3 * C + k4 * L
    return R

def expo_linear_growth(t, B, alpha, gamma, t_switch, k1, k2, k3, k4):
    idx = min(int(t), len(T) - 1)  # Dentro de los limites del √≠ndice
    R = recursos(T[idx], M[idx], C[idx], L[idx], k1, k2, k3, k4)

    if t < t_switch:
        return alpha * R * B  # Fase exponencial (dB/dt = alpha * R * B)
    else:
        return gamma * R      # Fase lineal (dB/dt = gamma * R)
```

Para este modelo expolineal definiremos los siguientes parametros, los cuales impactan severamente en las predicciones. Los cuales son suceptibles a mejora.

```python
# Par√°metros iniciales
alpha = 0.2  # Tasa de crecimiento exponencial
gamma = 0.05  # Tasa de crecimiento lineal
t_switch = 3  # Tiempo de transici√≥n de un comportamiento exponencial a lineal (√≠ndice)
k1, k2, k3, k4 = 0.1, 0.2, 0.05, 0.15  # Coeficientes de la funci√≥n de recursos

# Los coeficientes de la funci√≥n de recursos se ajustan de acuerdo a las descripciones
# de la incidencia de las variables en el crecimiento del tomate cherry

# Condici√≥n de biomasa inicial
B0 = [1.0]

# Intervalo de tiempo
t_span = (0, len(T) - 1)
```

Finalmente hacemos la resoluci√≥n del modelo de EDO planteado

```python
# Resoluci√≥n de la la EDO planteada
sol = solve_ivp(expo_linear_growth, t_span, B0, args=(alpha, gamma, t_switch, k1, k2, k3, k4), t_eval=np.arange(len(T)))

# Soluci√≥n grafica
plt.plot(sol.t, sol.y[0], label='Biomasa simulada')
plt.xlabel('Tiempo')
plt.ylabel('Biomasa')
plt.title('Simulaci√≥n del crecimiento expo-lineal')
plt.legend()
plt.show()
```

#### 5.3.1 Simulaci√≥n de datos

Dado que no contamos con medidas reales de la biomasa en tomates cherrys, generaremos un conjunto de datos sinteticos basados en el modelo expo-lineal que se ajusta m√°s a las descripciones descritas <a href="https://bio.libretexts.org/Bookshelves/Botany/Inanimate_Life_(Briggs)/01%3A_Chapters/1.25%3A_Plant_growthpatterns_limitations_and_models#title">Plant Growth Patterns , Limitations, and Models</a>. Este enfoque (expo-lineal) considera que las plantas atraviesan una fase de crecimiento exponencial seguida de una fase de crecimiento lineal, lo que lo hace adecuado para representar el desarrollo de la biomasa del tomate cherry.  

```python
# Generaramos datos sint√©ticos basados en el modelo propuesto expo-lineal (se asemeja al comportamiento descrito del crecimiento del Tomate Cherry)
def generar_datos_sinteticos(T, M, C, L, alpha, gamma, t_switch, k1, k2, k3, k4):
    t_span = (0, len(T) - 1)
    t_eval = np.arange(len(T))
    sol = solve_ivp(expo_linear_growth, t_span, B0, args=(alpha, gamma, t_switch, k1, k2, k3, k4), t_eval=np.arange(len(T)))
    return sol.t, sol.y[0]


def agr_ruido(datos, sigma=0.05):
    ruido = np.random.normal(0, sigma, size=len(datos))
    return datos + ruido
```

```python
data_sim = generar_datos_sinteticos(T,M,C,L,alpha,gamma,t_switch,k1,k2,k3,k4) # Retornamos tiempo y datos
data_ruido = agr_ruido(data_sim[1])
```

```python
def perturbar_datos(datos, intervalo=20, intensidad=0.3):
    datos_perturbados = datos.copy()
    for i in range(0, len(datos), intervalo):
        datos_perturbados[i] += np.random.uniform(-intensidad, intensidad)
    return datos_perturbados

data_pert = perturbar_datos(data_ruido,20,2.2)

plt.plot(data_sim[0], data_pert, label='Biomasa simulada')
plt.xlabel('Tiempo ')
plt.ylabel('Biomasa')
plt.title('Simulaci√≥n del crecimiento expo-lineal')
plt.legend()
plt.show()
```

#### 5.3.2 Ajuste de parametros

Optimizamos para encontrar los par√°metros que mejor ajusten la simulaci√≥n a los datos perturbados. Minimizamos la diferencia cuadr√°tica entre los valores simulados y los valores perturbados y comparamos los datos ajustados con los datos perturbados para evaluar el rendimiento del modelo.

```python
# Definir la funci√≥n de residuo (error cuadr√°tico entre la simulaci√≥n y los datos perturbados)
def residuo(parametros, T, M, C, L, datos_perturbados):
    # Extraer par√°metros
    alpha, gamma, t_switch, k1, k2, k3, k4 = parametros

    # Simular los datos con los par√°metros actuales
    t_span = (0, len(T) - 1)
    sol = solve_ivp(expo_linear_growth, t_span, [1.0], args=(alpha, gamma, t_switch, k1, k2, k3, k4), t_eval=np.arange(len(T)))

    # Comparar la simulaci√≥n con los datos perturbados
    simulacion = sol.y[0]
    error = np.sum((simulacion - datos_perturbados)**2)  # Error cuadr√°tico

    return error
```

```python
# Par√°metros iniciales para la optimizaci√≥n
parametros_iniciales = [0.1, 0.05, 3, 0.1, 0.2, 0.05, 0.15]

# Optimizaci√≥n: minimizar el error cuadr√°tico
resultado = minimize(residuo, parametros_iniciales, args=(T, M, C, L, data_pert), bounds=[(0, 3), (0, 3), (0, 20), (0, 3), (0, 3), (0, 3), (0, 3)],method='Powell')

# Par√°metros ajustados
print("Par√°metros ajustados:", resultado.x)

# Recuperar los par√°metros ajustados
alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt = resultado.x

# Generar la simulaci√≥n con los par√°metros ajustados
sol = solve_ivp(expo_linear_growth, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt), t_eval=np.arange(len(T)))

# Graficar comparaci√≥n entre los datos simulados perturbados y la simulaci√≥n ajustada
plt.plot(data_sim[0],data_sim[1], label='Biomasa simulada')
plt.plot(sol.t, sol.y[0], label='Simulaci√≥n de la biomasa ajustada')
plt.xlabel('Tiempo ')
plt.ylabel('Biomasa ')
plt.legend()
plt.title('Ajuste de par√°metros al modelo expo-lineal con perturbaciones')
plt.show()
```

```python

def expo_linear_growth_mod(t, B, alpha, gamma, t_switch, k1, k2, k3, k4, T, M, C, L):
    idx = min(int(t), len(T) - 1)  # Dentro de los limites del √≠ndice
    R = recursos(T[idx], M[idx], C[idx], L[idx], k1, k2, k3, k4)

    if t < t_switch:
        return alpha * R * B  # Fase exponencial (dB/dt = alpha * R * B)
    else:
        return gamma * R

def aplicar_variacion(vector, porcentaje):
    """
    Aplica una variaci√≥n porcentual a un vector.
    :param vector: Vector original (1 x N).
    :param porcentaje: Porcentaje de variaci√≥n (positivo o negativo).
    :return: Vector modificado.
    """
    factor = 1 + porcentaje / 100  # Convertir porcentaje a factor multiplicativo
    return vector * factor

# Definir el porcentaje de variaci√≥n deseado
porcentaje_variacion = 30

# Aplicar variaciones a cada variable
T_mas = aplicar_variacion(T, porcentaje_variacion)  # +30%
T_menos = aplicar_variacion(T, -porcentaje_variacion)  # -30%

M_mas = aplicar_variacion(M, porcentaje_variacion)  # +30%
M_menos = aplicar_variacion(M, -porcentaje_variacion)  # -30%

C_mas = aplicar_variacion(C, porcentaje_variacion)  # +30%
C_menos = aplicar_variacion(C, -porcentaje_variacion)  # -30%

L_mas = aplicar_variacion(L, porcentaje_variacion)  # +30%
L_menos = aplicar_variacion(L, -porcentaje_variacion)  # -30%

resultados = []
etiquetas = []

B0 = [1.0]
t_span = (0, len(T) - 1)

# Simulaciones para temperatura

sol_temp_mas = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T_mas, M, C, L), t_eval=np.arange(len(T)))
resultados.append(sol_temp_mas.y[0])
etiquetas.append(f'Temperatura +{porcentaje_variacion}%')

sol_temp_menos = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T_menos, M, C, L), t_eval=np.arange(len(T)))
resultados.append(sol_temp_menos.y[0])
etiquetas.append(f'Temperatura -{porcentaje_variacion}%')

# Simulaciones para humedad
sol_hum_mas = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T, M_mas, C, L), t_eval=np.arange(len(T)))
resultados.append(sol_hum_mas.y[0])
etiquetas.append(f'Humedad +{porcentaje_variacion}%')

sol_hum_menos = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T, M_menos, C, L), t_eval=np.arange(len(T)))
resultados.append(sol_hum_menos.y[0])
etiquetas.append(f'Humedad -{porcentaje_variacion}%')

# Simulaciones para conductividad el√©ctrica
sol_con_mas = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T, M, C_mas, L), t_eval=np.arange(len(T)))
resultados.append(sol_con_mas.y[0])
etiquetas.append(f'Conductividad +{porcentaje_variacion}%')

sol_con_menos = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt,  T, M, C_menos, L), t_eval=np.arange(len(T)))
resultados.append(sol_con_menos.y[0])
etiquetas.append(f'Conductividad -{porcentaje_variacion}%')

# Simulaciones para luminosidad solar
sol_lu_mas = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt, T, M, C, L_mas), t_eval=np.arange(len(T)))
resultados.append(sol_lu_mas.y[0])
etiquetas.append(f'Luminosidad +{porcentaje_variacion}%')

sol_lu_menos = solve_ivp(expo_linear_growth_mod, (0, len(T)-1), [1.0], args=(alpha_opt, gamma_opt, t_switch_opt, k1_opt, k2_opt, k3_opt, k4_opt,  T, M, C, L_menos), t_eval=np.arange(len(T)))
resultados.append(sol_lu_menos.y[0])
etiquetas.append(f'Luminosidad -{porcentaje_variacion}%')

# Graficar los resultados
plt.figure(figsize=(12, 8))
for i, resultado in enumerate(resultados):
    plt.plot(resultado, label=etiquetas[i])

plt.title('Impacto de las Variaciones Ambientales en el Crecimiento de la Biomasa')
plt.xlabel('Tiempo')
plt.ylabel('Biomasa')
plt.legend()
plt.grid(True)
plt.show()
```
